```python
import numpy as np
import matplotlib.pyplot as plt

randArray = np.random.randint(low =0, high=100, size=20)
print(randArray)
```

```python
plt.scatter(range(len(randArray)),randArray)
plt.xlabel("Index")
plt.ylabel("Wert")
plt.show()
```

```python
def calc_mittelwert(liste):
    print(liste)
    summe=0

    for value in liste:
        summe+=value

    return summe/len(liste)
```

```python
mean = calc_mittelwert(randArray)
print(mean)

plt.scatter(range(len(randArray)),randArray)
plt.xlabel("Index")
plt.ylabel("Wert")
plt.plot([0, len(randArray)], [mean, mean], color="Red")
plt.show()
```

```python
def calc_median(liste):
    print(liste)
    sorted(liste)
    print(liste)

    medianIndex = int(len(liste)/2)
    print(medianIndex)
    return liste[medianIndex]
```

```python
mean = calc_mittelwert(randArray)
print(mean)

median = calc_median(randArray)
print(median)

plt.scatter(range(len(randArray)),randArray)
plt.xlabel("Index")
plt.ylabel("Wert")
plt.plot([0, len(randArray)], [mean, mean], color="Red")
plt.plot([0, len(randArray)], [median, median], color="Green")
plt.show()
```
