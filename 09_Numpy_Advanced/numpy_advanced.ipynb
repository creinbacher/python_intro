```python
import numpy as np

x = np.array([1,2,3,4,1000])
print(x)
print(x.shape)
print(type(x))
print(x.dtype)

x = x.astype(np.float64)
print(x)
print(x.dtype)

# weil 1000 nicht in int8 passt gibts einen overflow
x = x.astype(np.int8)
print(x)
print(x.dtype)
```

```python
myArray = np.arange(10)
print(myArray)

myArray2 = np.linspace(start=0.0, stop=10.0, num=125)
print(myArray2)

myArray3 = np.linspace(start=0.0, stop=10.0, num=125, dtype=np.int32)
print(myArray3)
```
