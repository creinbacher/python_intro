```python
myList =[]

for i in range(10):
    myList.append(i)

print(myList)
```

```python
#list comprehansion
myList = [i for i in range(10)]
print(myList)

myList2 = [i**2 for i in range(0,20,2)]
print(myList2)

myList3 = [i**2 for i in range(20) if i % 2 ==0]
print(myList3)

myList4 = [i**2 if i % 2 ==0 else -1 for i in range(20) ]
print(myList4)
```

```python
#                     START; STOP
myListSliced = myList4[:5]
print(myListSliced)

myListSliced = myList4[4:8]
print(myListSliced)
```

```python
import numpy as np
import matplotlib.pyplot as plt

myArray = np.arange(16).reshape((8,2))
print(myArray)

print(myArray[1:2])

#           jede zeile, spalte 0
x_coordinats = myArray[:,0]
#           jede zeile, spalte 1
y_coordinats = myArray[:,1]
print(x_coordinats)
print(y_coordinats)

plt.scatter(x_coordinats, y_coordinats)
plt.show()
```

```python
print(myArray)
x_coordinats = myArray[:,0].copy()
y_coordinats = myArray[:,1].copy()
# ohne das copy oben wuerde hier der speicher vom myArray ueberschrieben
x_coordinats[0] = -10
print(myArray)

plt.scatter(x_coordinats, y_coordinats)
plt.show()
```
