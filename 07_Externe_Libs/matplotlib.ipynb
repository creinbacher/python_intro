```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4])
y = np.array([3,45,12,32])
```

```python
plt.plot(x, y, color="red")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")
plt.show()
```
